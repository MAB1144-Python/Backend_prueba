```python
import pandas as pd
df_cl = pd.DataFrame([])
for cl in response['data']:
    df_cl1 = pd.DataFrame.from_dict(cl['quotes'])
    df_cl1['ID_cliente'] = cl['id']
    df_cl1['description'] = cl['description']
    df_cl1['active'] = cl['active']
    df_cl1['birthDate'] = cl['birthDate']
    df_cl1['sex'] = cl['sex']
    df_cl1['admissionDate'] = cl['admissionDate']
    df_cl = pd.concat([df_cl,df_cl1])

df_cl
```

```python

import numpy as np
filas = pd.DataFrame([],columns=['ID_cliente', 'description', 'active', 'birthDate', 'sex', 'admissionDate','id', 'date', 'status', 'cup_id','modulo','pregunta','respuesta'])
print(df_cl.keys())
for cl in range(len(df_cl)):
    info1 = df_cl.iloc[cl][['ID_cliente', 'description', 'active', 'birthDate', 'sex', 'admissionDate','id', 'date', 'status', 'cup_id']].values
    for mod in json.loads(df_cl.iloc[cl]['story']):
        for fl in mod['filas']:
            try:
                filas1 = pd.DataFrame([np.append(info1, [mod['mod'],fl['pregunta'],fl['respuesta']])],columns=['ID_cliente', 'description', 'active', 'birthDate', 'sex', 'admissionDate', 'id', 'date', 'status', 'cup_id','modulo','pregunta','respuesta'])
            except:
                filas1 = pd.DataFrame([np.append(info1, [mod['mod'],fl['pregunta'],None])],columns=['ID_cliente', 'description', 'active', 'birthDate', 'sex', 'admissionDate', 'id', 'date', 'status', 'cup_id','modulo','pregunta','respuesta'])
            filas = pd.concat([filas,filas1])
print(filas)        
```
